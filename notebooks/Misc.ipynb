```python
import pandas as pd 
filename1 = "/Users/shriramsunder/Downloads/MAF freelance - Done.csv"
filename2 = "/Users/shriramsunder/Downloads/MAF X SD (2).xlsx"

done_df = pd.read_csv(filename1)
maf_df = pd.read_excel(filename2)
```

```python
merged_df = pd.merge(maf_df[maf_df["Content Writer"] == "SA"], done_df, on="Product Name (Shared by Agency)", how="left", indicator=True)
```

```python
rel_merged_df = merged_df.loc[merged_df["_merge"] == "left_only", [col for col in merged_df.columns if "_y" not in col]]
```

```python
rel_merged_df.columns = [col.replace("_x", "") for col in rel_merged_df.columns]
```

```python
rel_merged_df.to_excel("rows_in_massive_file_not_found_in_done.xlsx")
```
